Dirty data can also ocurr when collection information from different sources. If these sources use different conventions to represent such data, additional efforts are required to homogenize it at a later stage.

```python
def unify_victoria_secret(df):
    """
    We want that all brands that are related to Victoria's Secret
    have `victoria's secret` as their brand instead of what they
    currently have.
    """
    df = df.copy()
    new_string = "victoria's secret"
    # Write your code here

    currently = [
        "Victorias-Secret",
        "Victoria's Secret",
        "Victoria's Secret Pink",
        "VS",
        "VSecret",
    ]

    # Reemplazar las variantes por el nuevo nombre
    df['brand_name'] = df['brand_name'].replace(currently, new_string, regex=False)

    return df
```

```python
# Función para clasificar valores en floats o strings
def count_types(price):
    float_count = 0
    string_count = 0

    for value in price:
        # Verificar si el valor es float
        if isinstance(value, float):
            float_count += 1
        # Verificar si el valor puede convertirse a float
        elif isinstance(value, str):
            try:
                float(value)  # Intentar convertirlo
                float_count += 1
            except ValueError:
                string_count += 1
        else:
            string_count += 1  # Contar valores no convertibles como strings

    return float_count, string_count

# Contar floats y strings en la columna "price"
floats, strings = count_types(df['price'])

# Imprimir resultados
print(f"Cantidad de floats: {floats}")
print(f"Cantidad de strings: {strings}")
```

```python
# Función para verificar si el valor es un string
def is_string(price):
    return isinstance(price, str)

# Función para normalizar precios
def normalize_price(price):
    if isinstance(price, str):
        # Eliminar 'USD' si está presente
        price = price.replace('USD', '').strip()
        # Eliminar '$' si está presente
        price = price.replace('$', '').strip()
        # Manejar rangos de precios
        if '-' in price or '–' in price:
            prices = price.replace('–', '-').split('-')
            price = min(prices, key=lambda x: float(x.strip().replace('.', '').replace(',', '.')))
        # Reemplazar ',' por '' y '.' por ',' si ',' está presente en la cadena
        if ',' in price:
            price = price.replace('.', '').replace(',', '.')
        # Convertir a float
        return float(price)
    return price

# Identificar valores que son strings
df['es_string'] = df['price'].apply(is_string)

# Normalizar precios
df['precios_normalizados'] = df['price'].apply(normalize_price)

print(df)
```
